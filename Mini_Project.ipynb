```python
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Read the cleaned CSV file
data = pd.read_csv('/content/cleaned_data.csv')

# Step 2: Import pyplot and matplotlib

# 1. Distribution of math scores (Bar graph)
plt.hist(data['math score'], bins=10)
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.title('Distribution of Math Scores')
plt.show()

# 2. Distribution of reading scores (Bar graph)
plt.hist(data['reading score'], bins=10)
plt.xlabel('Reading Score')
plt.ylabel('Frequency')
plt.title('Distribution of Reading Scores')
plt.show()

# 3. Distribution of writing scores (Line plot)
plt.plot(data['writing score'])
plt.xlabel('Student')
plt.ylabel('Writing Score')
plt.title('Distribution of Writing Scores')
plt.show()

# 4. Average math score by race/ethnicity (Pie chart)
avg_math_by_group = data.groupby('race/ethnicity')['math score'].mean()
plt.pie(avg_math_by_group, labels=avg_math_by_group.index, autopct='%1.1f%%')
plt.title('Average Math Score by Race/Ethnicity')
plt.show()

# 5. Average reading score by race/ethnicity (Histogram)
avg_reading_by_group = data.groupby('race/ethnicity')['reading score'].mean()
plt.hist(avg_reading_by_group, bins=5)
plt.xlabel('Average Reading Score')
plt.ylabel('Frequency')
plt.title('Distribution of Average Reading Scores by Race/Ethnicity')
plt.show()

# 6. Average writing score by race/ethnicity (Line plot)
avg_writing_by_group = data.groupby('race/ethnicity')['writing score'].mean()
plt.plot(avg_writing_by_group)
plt.xlabel('Race/Ethnicity')
plt.ylabel('Average Writing Score')
plt.title('Average Writing Score by Race/Ethnicity')
plt.show()

# 7. Average math score by parental education (Histogram)
avg_math_by_education = data.groupby('parental level of education')['math score'].mean()
plt.hist(avg_math_by_education, bins=5)
plt.xlabel('Average Math Score')
plt.ylabel('Frequency')
plt.title('Distribution of Average Math Scores by Parental Education')
plt.xticks(rotation=45)
plt.show()

# 8. Average reading score by parental education (Bar plot)
avg_reading_by_education = data.groupby('parental level of education')['reading score'].mean()
plt.bar(avg_reading_by_education.index, avg_reading_by_education.values)
plt.xlabel('Parental Education')
plt.ylabel('Average Reading Score')
plt.title('Average Reading Score by Parental Education')
plt.xticks(rotation=45)
plt.show()

# 9. Average writing score by parental education (Pie chart)
avg_writing_by_education = data.groupby('parental level of education')['writing score'].mean()
plt.pie(avg_writing_by_education, labels=avg_writing_by_education.index, autopct='%1.1f%%')
plt.title('Average Writing Score by Parental Education')
plt.show()

# 10. Average scores for students who completed the test preparation course versus those who didn't (Histogram)
completed_scores = data[data['test preparation course'] == 'completed'][['math score', 'reading score', 'writing score']].mean()
not_completed_scores = data[data['test preparation course'] == 'none'][['math score', 'reading score', 'writing score']].mean()

plt.hist([completed_scores, not_completed_scores], bins=5, label=['Completed', 'Not Completed'], color=['blue', 'orange'], alpha=0.7)
plt.xlabel('Average Score')
plt.ylabel('Frequency')
plt.title('Distribution of Average Scores by Test Preparation Course')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Read the cleaned CSV file
data = pd.read_csv('/content/cleaned_data.csv')

# Step 2: Create a dashboard with subplots
fig, axs = plt.subplots(3, 3, figsize=(12, 12))
fig.suptitle('Student Performance Dashboard')

# 1. Distribution of math scores (Bar graph)
axs[0, 0].hist(data['math score'], bins=10)
axs[0, 0].set_xlabel('Math Score')
axs[0, 0].set_ylabel('Frequency')
axs[0, 0].set_title('Distribution of Math Scores')

# 2. Distribution of reading scores (Bar graph)
axs[0, 1].hist(data['reading score'], bins=10)
axs[0, 1].set_xlabel('Reading Score')
axs[0, 1].set_ylabel('Frequency')
axs[0, 1].set_title('Distribution of Reading Scores')

# 3. Distribution of writing scores (Line plot)
axs[0, 2].plot(data['writing score'])
axs[0, 2].set_xlabel('Student')
axs[0, 2].set_ylabel('Writing Score')
axs[0, 2].set_title('Distribution of Writing Scores')

# 4. Average math score by race/ethnicity (Pie chart)
avg_math_by_group = data.groupby('race/ethnicity')['math score'].mean()
axs[1, 0].pie(avg_math_by_group, labels=avg_math_by_group.index, autopct='%1.1f%%')
axs[1, 0].set_title('Average Math Score by Race/Ethnicity')

# 5. Average reading score by race/ethnicity (Histogram)
avg_reading_by_group = data.groupby('race/ethnicity')['reading score'].mean()
axs[1, 1].hist(avg_reading_by_group, bins=5)
axs[1, 1].set_xlabel('Average Reading Score')
axs[1, 1].set_ylabel('Frequency')
axs[1, 1].set_title('Distribution of Average Reading Scores by Race/Ethnicity')

# 6. Average writing score by race/ethnicity (Line plot)
avg_writing_by_group = data.groupby('race/ethnicity')['writing score'].mean()
axs[1, 2].plot(avg_writing_by_group)
axs[1, 2].set_xlabel('Race/Ethnicity')
axs[1, 2].set_ylabel('Average Writing Score')
axs[1, 2].set_title('Average Writing Score by Race/Ethnicity')

# 7. Average math score by parental education (Histogram)
avg_math_by_education = data.groupby('parental level of education')['math score'].mean()
axs[2, 0].hist(avg_math_by_education, bins=5)
axs[2, 0].set_xlabel('Average Math Score')
axs[2, 0].set_ylabel('Frequency')
axs[2, 0].set_title('Distribution of Average Math Scores by Parental Education')
axs[2, 0].tick_params(axis='x', rotation=45)

# 8. Average reading score by parental education (Bar plot)
avg_reading_by_education = data.groupby('parental level of education')['reading score'].mean()
axs[2, 1].bar(avg_reading_by_education.index, avg_reading_by_education.values)
axs[2, 1].set_xlabel('Parental Education')
axs[2, 1].set_ylabel('Average Reading Score')
axs[2, 1].set_title('Average Reading Score by Parental Education')
axs[2, 1].tick_params(axis='x', rotation=45)

# 9. Average writing score by parental education (Pie chart)
avg_writing_by_education = data.groupby('parental level of education')['writing score'].mean()
axs[2, 2].pie(avg_writing_by_education, labels=avg_writing_by_education.index, autopct='%1.1f%%')
axs[2, 2].set_title('Average Writing Score by Parental Education')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the dashboard
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import pandas as pd

# Step 1: Read the cleaned CSV file
data = pd.read_csv('/content/cleaned_data.csv')

# Step 2: Create the dashboard layout
fig = plt.figure(figsize=(16, 12))
fig.suptitle('Student Performance Dashboard', fontsize=16)

# Gridspec configuration
gs = gridspec.GridSpec(3, 3)

# Plot 1: Distribution of math scores (Bar graph)
ax1 = fig.add_subplot(gs[0, 0])
ax1.hist(data['math score'], bins=10, color='#007acc')
ax1.set_xlabel('Math Score')
ax1.set_ylabel('Frequency')
ax1.set_title('Distribution of Math Scores')

# Plot 2: Distribution of reading scores (Bar graph)
ax2 = fig.add_subplot(gs[0, 1])
ax2.hist(data['reading score'], bins=10, color='#d9534f')
ax2.set_xlabel('Reading Score')
ax2.set_ylabel('Frequency')
ax2.set_title('Distribution of Reading Scores')

# Plot 3: Distribution of writing scores (Line plot)
ax3 = fig.add_subplot(gs[0, 2])
ax3.plot(data['writing score'], color='#ffa500')
ax3.set_xlabel('Student')
ax3.set_ylabel('Writing Score')
ax3.set_title('Distribution of Writing Scores')

# Plot 4: Average math score by race/ethnicity (Pie chart)
ax4 = fig.add_subplot(gs[1, 0])
avg_math_by_group = data.groupby('race/ethnicity')['math score'].mean()
colors = ['#8dd3c7', '#ffffb3', '#bebada', '#fb8072', '#80b1d3']
ax4.pie(avg_math_by_group, labels=avg_math_by_group.index, autopct='%1.1f%%', colors=colors)
ax4.set_title('Average Math Score by Race/Ethnicity')

# Plot 5: Average reading score by race/ethnicity (Histogram)
ax5 = fig.add_subplot(gs[1, 1])
avg_reading_by_group = data.groupby('race/ethnicity')['reading score'].mean()
ax5.bar(avg_reading_by_group.index, avg_reading_by_group.values, color='#ffbb78')
ax5.set_xlabel('Race/Ethnicity')
ax5.set_ylabel('Average Reading Score')
ax5.set_title('Average Reading Score by Race/Ethnicity')

# Plot 6: Average writing score by race/ethnicity (Line plot)
ax6 = fig.add_subplot(gs[1, 2])
ax6.plot(avg_writing_by_group, color='#9467bd')
ax6.set_xlabel('Race/Ethnicity')
ax6.set_ylabel('Average Writing Score')
ax6.set_title('Average Writing Score by Race/Ethnicity')

# Plot 7: Average math score by parental education (Histogram)
ax7 = fig.add_subplot(gs[2, 0])
avg_math_by_education = data.groupby('parental level of education')['math score'].mean()
ax7.bar(avg_math_by_education.index, avg_math_by_education.values, color='#98abc5')
ax7.set_xlabel('Parental Education')
ax7.set_ylabel('Average Math Score')
ax7.set_title('Average Math Score by Parental Education')
ax7.tick_params(axis='x', rotation=45)

# Plot 8: Average reading score by parental education (Bar plot)
ax8 = fig.add_subplot(gs[2, 1])
avg_reading_by_education = data.groupby('parental level of education')['reading score'].mean()
ax8.bar(avg_reading_by_education.index, avg_reading_by_education.values, color='#c5b0d5')
ax8.set_xlabel('Parental Education')
ax8.set_ylabel('Average Reading Score')
ax8.set_title('Average Reading Score by Parental Education')
ax8.tick_params(axis='x', rotation=45)

# Plot 9: Average writing score by parental education (Pie chart)
ax9 = fig.add_subplot(gs[2, 2])
avg_writing_by_education = data.groupby('parental level of education')['writing score'].mean()
ax9.pie(avg_writing_by_education, labels=avg_writing_by_education.index, autopct='%1.1f%%', colors=colors)
ax9.set_title('Average Writing Score by Parental Education')


# Adjust spacing between subplots
fig.tight_layout()

# Display the dashboard
plt.show()
```
